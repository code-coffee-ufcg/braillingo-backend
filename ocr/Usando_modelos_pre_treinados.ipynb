## Utilizando um modelo pré-treinado de reconhecimento de palavras e números 

Primeiramente devemos importar a biblioteca de optical caracter recognition keras_ocr e a biblioteca matplotlib para plotar os resultados.

```python
import keras_ocr
import matplotlib.pyplot as plt
```

Modelo pré-treinado pipeline e configurado com os pesos do detector e reconhecedor.

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

Lendo uma lista de imagens pela função tools.read

```python
imagens = [
           keras_ocr.tools.read(img) for img in ['/content/drive/MyDrive/Colab Notebooks/imagens/outdoor.jpeg',
                                                  '/content/drive/MyDrive/Colab Notebooks/imagens/copo_personalizado.jpg'
           ]
]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,20))
axs[0].imshow(imagens[0])
axs[1].imshow(imagens[1])
```

Realizando o reconhecimento e predição.

```python
predicao = pipeline.recognize(imagens)
```

Adicionando as anotações as imagens e plotando os resultados.

```python
fig, axs = plt.subplots(nrows=len(imagens), figsize=(20, 20))
for ax, image, predictions in zip(axs, imagens, predicao):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
